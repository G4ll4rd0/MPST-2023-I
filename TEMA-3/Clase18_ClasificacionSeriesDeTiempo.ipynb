# ¿Por qué la clasificación de series temporales?

En primer lugar, es importante subrayar por qué este problema es tan importante hoy en día y, por lo tanto, por qué es muy interesante comprender el papel y el potencial del aprendizaje profundo en este sector.

Durante los últimos años, la Clasificación de Series Temporales se ha convertido en uno de los problemas más desafiantes en la Ciencia de Datos. Esto ha sucedido porque cualquier problema de clasificación que utilice datos teniendo en cuenta alguna noción de clasificación puede tratarse como un problema de clasificación de series temporales.

Las series temporales están presentes en muchas aplicaciones del mundo real que van desde el 
- cuidado de la salud,
- el reconocimiento de la actividad humana,
- la ciberseguridad,
- las finanzas,
- el marketing,
- la detección automatizada de enfermedades,
- la detección de anomalías, etc.

Dado que la disponibilidad de datos temporales ha aumentado significativamente en los últimos años, muchas áreas se están interesando mucho en las aplicaciones basadas en series temporales, y luego se han propuesto muchos algoritmos nuevos.

Todos estos algoritmos, además de los basados en <font color=red>**aprendizaje profundo, requieren algún tipo de ingeniería de características**</font> como una tarea separada antes de realizar la clasificación, y esto puede implicar la pérdida de información y el aumento del tiempo de desarrollo. Por el contrario, los modelos de aprendizaje profundo ya incorporan este tipo de ingeniería de funciones internamente, optimizándola y eliminando la necesidad de hacerlo manualmente. Por tanto, <font color=blue>**son capaces de extraer información de la serie temporal de una forma más rápida, directa y completa.**</font>

Para ilustrar esta nueva metodología vamos a introducir el problema real de <font color=red>**clasificación de series temporales de reconocimiento de actividad (HAR), por sus siglas en inglés,**</font> a partir de datos multivariantes de acelerómetros grabados desde un smartphone, y en cómo desarrollar modelos de clasificación de aprendizaje automático y aprendizaje profundo para abordar el problema.

## 1. Human activity recognition (HAR)

Se trata de un problema difícil, ya que no hay formas obvias o directas de relacionar los datos registrados de los sensores con actividades humanas concretas y cada sujeto puede realizar una actividad con variaciones significativas, lo que da lugar a variaciones en los datos registrados de los sensores. El objetivo es registrar los datos de los sensores y las actividades correspondientes de sujetos específicos, ajustar un modelo a partir de estos datos y generalizar el modelo para clasificar la actividad de nuevos sujetos no observados a partir de sus datos de sensores.

Debido a esta variación grande que puede existir entre las diversas señales para realizar clasificación, se observa el potencial que tienen los algoritmos de deep learning para extraer las características más complejas de este tipo de señales y en esencia poder hacer el proceso de clasificación menos costoso y escalable.

## 2. Aprendizaje supervisado Representación de datos

Una ventana de datos dada puede tener múltiples variables, como los ejes x, y, y z de un sensor acelerómetro. Concretemos esto con un ejemplo. Tenemos datos del sensor durante 10 minutos; que puede tener el siguiente aspecto:
```
x,    y,   z,  activity
1.1, 2.1, 0.1, 1
1.2, 2.2, 0.2, 1
1.3, 2.3, 0.3, 1
...
```
Si los **datos se registran a 8 Hz**, eso significa que habrá ocho filas de datos para un segundo de tiempo transcurrido realizando una actividad. Podemos elegir que una **ventana de datos represente un segundo** de datos; eso significa ocho filas de datos para un sensor de 8 Hz. Si tenemos datos x, y, y z, eso significa que tendríamos 3 variables. Por lo tanto, una sola ventana de datos sería una matriz bidimensional con ocho pasos de tiempo y tres características. Una ventana representaría una muestra. Un minuto de datos representaría 480 puntos de datos del sensor, o 60 ventanas de ocho pasos temporales. El total de 10 minutos de datos representaría 4.800 puntos de datos, o 600 ventanas de datos.

Como hemos venido trabajando hasta ahora, necesitamos expresar nuestros datos como `número de muestras`, `el número de pasos temporales de una ventana` y el `número de características`. Para el ejemplo discutido anteriormente nuestro tensor tendría la siguiente dimensión
```
[600, 8, 3]
```
Observe que aumentando el tamaño de la ventana deslizante (número de pasos temporales) aumentaría el cómputo necesario para dicha red, pero a su vez se podría reconocer actividades más complejas. Para casos prácticos, es común usar uno o dos segundos de datos del sensor para clasificar dicho fragmento de actividad.

En estudios cómo *Window Size Impact in Human Activity Recognition, 2014* se menciona la siguiente:
> *Ventanas reducidas (2s o menos) demostraron proveer la mayor precisión de detección de actividad. De hecho, el mejor clasificador fue obtenido para ventanas muy cortas (0.25-0.5 s).*.

### 5.2 Cargar el dataset

Vamos a crear las funciones correspondientes para poder leer la data que necesitamos y usaremos para modelar.

````python
# Importar librerías
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Crear función que lee los datos de un archivo en formato txt y los retorna en un dataframe. Están separados por espacios en blanco
def load_file(filename):
    df = pd.read_csv(filename, delim_whitespace=True, header=None)
    return df.values
````

Solo usaremos los datos `total_acc_i_train.txt` para entrenar y `total_acc_i_test.txt` para probar. Para ello, crearemos una función que nos permita leer los datos de entrenamiento y de prueba.


````python
# Explicar de forma sencilla np.dstack
a = np.array([[1, 2, 3], [1, 2, 3]])
b = np.array([[4, 5, 6], [4, 5, 6]])
c = np.array([[7, 8, 9], [7, 8, 9]])

np.dstack((a, b, c)), np.dstack((a, b, c)).shape
````

````python
load_file('UCI HAR Dataset/train/Inertial Signals/body_acc_x_train.txt').shape
````

````python
# Crear función que lee los datos de entrenamiento y de prueba
def load_group(filenames, prefix=''):
    loaded = [load_file(prefix + f) for f in filenames]

    # Apilar los datos en un solo array de 3 dimensiones
    loaded = np.dstack(loaded)
    return loaded
````

````python
# Leer los datos de entrenamiento y de prueba
total_acc_train = load_group(['body_acc_x_train.txt'], 'UCI HAR Dataset/train/Inertial Signals/')
total_acc_test = load_group(['body_acc_x_test.txt'], 'UCI HAR Dataset/test/Inertial Signals/')

total_acc_train.shape, total_acc_test.shape
````

````python
# Crear función que lee los datos de entrenamiento y de prueba
# load a dataset group, such as train or test
def load_dataset(group, prefix=''):
    filepath = prefix + group + '/Inertial Signals/'
    # cargar los 9 archivos como un único array
    filenames = list()
    # total acceleration
    filenames += ['total_acc_x_' + group + '.txt', 'total_acc_y_' + group + '.txt',
                  'total_acc_z_' + group + '.txt']
    # body acceleration
    filenames += ['body_acc_x_' + group + '.txt', 'body_acc_y_' + group + '.txt',
                  'body_acc_z_' + group + '.txt']
    # body gyroscope
    filenames += ['body_gyro_x_' + group + '.txt', 'body_gyro_y_' + group + '.txt',
                  'body_gyro_z_' + group + '.txt']
    # cargar input data
    X = load_group(filenames, filepath)
    # cargar class output
    y = load_file(prefix + group + '/y_' + group + '.txt')
    return X, y
````

````python
# load all train
trainX, trainy = load_dataset('train', 'UCI HAR Dataset/')
print(trainX.shape, trainy.shape)
# load all test
testX, testy = load_dataset('test', 'UCI HAR Dataset/')
print(testX.shape, testy.shape)
````

````python
# resumir el balance de clases en una columna variable de salida usando DataFrame.value_counts()
def class_breakdown(data):
    df = pd.DataFrame(data)
    vc = df.value_counts() / len(data) * 100
    vc = vc.to_frame()
    vc.index.name = 'Class'
    vc.columns = ['n']
    vc['%'] = vc['n'].round(2)
    print('balance de clases', vc, sep='\n')
    # Graficar el balance de clases
    vc['%'].plot(kind='bar')
    plt.show()
````

````python
print('train Data')
class_breakdown(trainy)

print('test Data')
class_breakdown(testy)

print('all Data')
class_breakdown(np.vstack([trainy, testy]))
````

### 5.3 Graficar serie de tiempo

Los datos brutos se componen de ventanas de datos de series temporales por variable, y las ventanas se solapan en un 50%. Esto sugiere que podemos ver alguna repetición en las observaciones como un gráfico de líneas a menos que se elimine el solapamiento. Podemos empezar cargando el conjunto de datos de entrenamiento utilizando las funciones desarrolladas anteriormente

````python
# Cargar datos
trainX, trainy
````

````python
# load subject mapping
sub_map = load_file('UCI HAR Dataset/train/subject_train.txt')
train_subjects = np.unique(sub_map)
print(train_subjects)
````

````python
# Obtener todos los datos para un sujeto
def data_for_subject(X, y, sub_map, sub_id):
    # Obtener los índices de fila para el id del sujeto
    ix = [i for i in range(len(sub_map)) if sub_map[i]==sub_id]
    # Devolver solo los datos para el sujeto
    return X[ix, :, :], y[ix]
````

Ahora que tenemos los datos de un sujeto, podemos representarlos gráficamente. Los datos se componen de ventanas superpuestas. Podemos escribir una función para eliminar esta superposición y aplastar las ventanas para una variable dada en una larga secuencia que se puede trazar directamente como un gráfico de líneas.

````python
# Convertir una ventana de datos en una serie de tiempo
def to_series(windows):
    series = list()
    for window in windows:
        # Remover la superposición de la ventana
        half = int(len(window) / 2) - 1
        for value in window[-half:]:
            series.append(value)
    return series
````

````python
# Graficar los datos de un sujeto
def plot_subject(X, y):
    plt.figure(figsize=(10, 8))
    # Determinar el número total de gráficos
    n, off = X.shape[2] + 1, 0
    # plot total acc
    for i in range(3):
        plt.subplot(n, 1, off+1)
        plt.plot(to_series(X[:, :, off]))
        plt.title('total acc '+str(i), y=0, loc='left', size=7)
        # Quitar los ticks para eliminar el desorden
        plt.yticks([])
        plt.xticks([])
        off += 1

    # plot body acc
    for i in range(3):
        plt.subplot(n, 1, off+1)
        plt.plot(to_series(X[:, :, off]))
        plt.title('body acc '+str(i), y=0, loc='left', size=7)
        # Quitar los ticks para eliminar el desorden
        plt.yticks([])
        plt.xticks([])
        off += 1
    # plot body gyro
    for i in range(3):
        plt.subplot(n, 1, off+1)
        plt.plot(to_series(X[:, :, off]))
        plt.title('body gyro '+str(i), y=0, loc='left', size=7)
        # Quitar los ticks para eliminar el desorden
        plt.yticks([])
        plt.xticks([])
        off += 1
    # plot activities
    plt.subplot(n, 1, n)
    plt.plot(y)
    plt.title('activity', y=0, loc='left', size=7)
    # Quitar los ticks para eliminar el desorden
    plt.yticks([])
    plt.xticks([])
    plt.show()
````

````python
# Obtener todos los datos para un sujeto
sub_id = train_subjects[1]
print('sujeto', sub_id)

trainX_sub, trainy_sub = data_for_subject(trainX, trainy, sub_map, sub_id)
print(trainX_sub.shape, trainy_sub.shape)

# Graficar los datos de un sujeto
plot_subject(trainX_sub, trainy_sub)
````

En el gráfico, podemos ver periodos de gran movimiento correspondientes a las actividades 1, 2 y 3: las actividades de caminar. También podemos ver mucha menos actividad (es decir, una línea relativamente recta) para las actividades con números más altos, 4, 5 y 6 (sentado, de pie y tumbado). Esto es una buena confirmación de que hemos cargado correctamente el conjunto de datos sin procesar. Podemos ver que este sujeto ha realizado la misma secuencia general de actividades dos veces, y algunas actividades se realizan más de dos veces. Esto sugiere que, para un sujeto determinado, no debemos hacer suposiciones sobre las actividades que puede haber realizado o su orden.

### 5.4 Graficar histogramas de frecuencia por sujeto

Tal y como está planteado el problema, nos interesa utilizar los datos de movimiento de algunos sujetos para predecir actividades a partir del movimiento de otros sujetos. Esto sugiere que debe haber regularidad en los datos de movimiento de los sujetos.


````python
# Graficar histogramas de frecuencia por sujeto
def plot_subject_histograms(X, y, sub_map, offset, n=10, title=''):

    fig = plt.figure(figsize=(10, 8))
    # poner supertitulos para los subplots
    fig.suptitle(title)

    # Obtenemos los ids de los sujetos
    subject_ids = np.unique(sub_map[:,0])
    # Enumeramos los sujetos
    for k in range(n):
        sub_id = subject_ids[k]
        # Obtenemos los datos para el sujeto
        subX, _ = data_for_subject(X, y, sub_map, sub_id)

        # total acc
        for i in range(3):
            ax = plt.subplot(n, 1, k+1)
            ax.set_xlim(-1,1)
            ax.hist(to_series(subX[:,:,offset+i]), bins=100)
            plt.yticks([])
            plt.xticks([-1,0,1])
    # plt.title(title)
    plt.show()
````

````python
# plot total acceleration histograms for subjects
plot_subject_histograms(trainX, trainy, sub_map, 0, title='total acc')
# plot body acceleration histograms for subjects
plot_subject_histograms(trainX, trainy, sub_map, 3, title='body acc')
# plot gyroscopic histograms for subjects
plot_subject_histograms(trainX, trainy, sub_map, 6, title='body gyro')

````

### 5.5 Graficar histogramas de frecuencia por actividad

También nos interesa ver si hay regularidad en los datos de movimiento de las actividades. Esto nos ayudará a determinar si podemos usar los datos de movimiento de una actividad para predecir otra actividad.

````python
# group data by activity
def data_by_activity(X, y, activities):
    # group windows by activity
    return {a:X[y[:,0]==a, :, :] for a in activities}

# plot histograms for each activity for a subject
def plot_activity_histograms(X, y, offset, title=''):
    # get a list of unique activities for the subject
    activity_ids = np.unique(y[:,0])
    # group windows by activity
    grouped = data_by_activity(X, y, activity_ids)


    fig = plt.figure(figsize=(10, 8))
    # poner supertitulos para los subplots
    fig.suptitle(title)
    for k in range(len(activity_ids)):
        act_id = activity_ids[k]
        # total acceleration
        for i in range(3):
            ax = plt.subplot(len(activity_ids), 1, k+1)
            ax.set_xlim(-1,1)
            # create histogra,
            plt.hist(to_series(grouped[act_id][:,:,offset+i]), bins=100)
            # create title
            plt.title('activity '+str(act_id), y=0, loc='left', size=10)
            # simplify axis
            plt.yticks([])
            plt.xticks([-1,0,1])
    plt.show()
````

````python
# plot total acceleration histograms per activity for a subject
plot_activity_histograms(trainX_sub, trainy_sub, 0, title='total acc')
# plot body acceleration histograms per activity for a subject
plot_activity_histograms(trainX_sub, trainy_sub, 3, title='body acc')
# plot gyroscopic histograms per activity for a subject
plot_activity_histograms(trainX_sub, trainy_sub, 6, title='body gyro')
````
